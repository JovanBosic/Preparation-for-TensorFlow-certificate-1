# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))   #fig size je dimenzija x*y (sirina * visina)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Two options for subplots

```python
# Subplot option 1
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10),np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quickly its okey to use pyplot method
* When plotting something more advanced use OO method

```python
over_50 = heart_disease[heart_disease.age>50]
over_50.head()
```

```python
len(over_50)
```

```python
len(heart_disease)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target');
```

```python
# OO method - object oriented method mixed with PYPLOT
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim([45,100]) # Promeni x osu od 45 do 100 da ide
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# Custumize the plot
ax.set(title='Heart Disease and Cholesterol Leveles',
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target"); # legend elements hvata iz c parametra 

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
over_50.target.values
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

# fig je cela figura a ax je samo jedan graf
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True) # onda je x osa ista kod oba

# Add data to ax0
scatter = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target)

# Customaze ax0
ax0.set(title="Heart Diese and Cholesterol Levels",
       #xlabel='Age',
       ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
# Add ameanline
ax0.axhline(y=over_50.chol.mean(),
           linestyle='--');

# add data to ax1
scatter=ax1.scatter(x=over_50.age,
                   y=over_50.thalach,
                   c=over_50.target)

# Customize ax1
ax1.set(title = "Heart Diesease and Max Heart Rate",
       xlabel = "age",
       ylabel="Max heart rate");
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title=" Target")
#add a mealine
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');

# add a title to the figure

fig.suptitle("Heart Disease Analyses", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter')# color map-change color schame
# Custumize the plot
ax.set(title='Heart Disease and Cholesterol Leveles',
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target"); # legend elements hvata iz c parametra 

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Customiying the y and x axis limitations

# Subplot of chol, age, thalach

# fig je cela figura a ax je samo jedan graf
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True) # onda je x osa ista kod oba

# Add data to ax0
scatter = ax0.scatter(x=over_50.age,
                     y=over_50.chol,
                     c=over_50.target,
                     cmap='winter')

# Customaze ax0
ax0.set(title="Heart Diese and Cholesterol Levels",
       #xlabel='Age',
       ylabel="Cholesterol")

# Change the x axis limits
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")
# Add ameanline
ax0.axhline(y=over_50.chol.mean(),
           linestyle='--');

# add data to ax1
scatter=ax1.scatter(x=over_50.age,
                   y=over_50.thalach,
                   c=over_50.target,
                   cmap='spring')

# Customize ax1
ax1.set(title = "Heart Diesease and Max Heart Rate",
       xlabel = "Age",
       ylabel="Max heart rate");

# Change ax1 x asix limit
# nije neophodno ako sharuju x osu
#ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title=" Target")
#add a mealine
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');

# add a title to the figure

fig.suptitle("Heart Disease Analyses", fontsize=16, fontweight="bold");
```
